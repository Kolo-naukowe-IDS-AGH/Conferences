## Przykłady
### Level 1 - problem regresji
### Regresja liniowa

### 1.1 Dane

```python
import numpy as np
from matplotlib import pyplot as plt

np.random.seed(12345)

X = np.arange(0, 20, 0.2)        #x: np.ndarray - 100-elementowy wektor 0, 0.2, 0.4 ... 19.6, 19.8
X
```

```python
X.shape
```

$y = ax + b$

```python
a = 0.7 
b = 1.3
y = a*X + b + np.random.randn(X. size) # funkcja limowa * szum

# wektor wejiciowy danych do modelu musi bye w postaci [[xl],[x2J,[x3],...J => wymiary: 1x100, zamiast 100x1 
X = X.reshape(-1,1)

plt.figure(figsize=(8, 8)) 
plt.plot(X, y, 'bo') 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show()
```

```python
f1 = lambda x: 0.8 * x + 1.35
f2 = lambda x: 0.9 * x + 0.9
f3 = lambda x: 0.6 * x + 1.5
plt.figure(figsize=(10, 10))
plt.plot(X, y, 'bo')
plt.plot ((0, 20), (f1(0), f1(20)), 'r-') 
plt.plot((0, 20), (f2(0), f2(20)), 'y-') 
plt.plot((0, 20), (f3(0), f3(20)), 'g-') 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show()
```

### 1.2 Regresja

```python
from sklearn.linear_model import LinearRegression
# tworzymy model dla regresji liniowej 
model = LinearRegression()
regression = model.fit(X,y)

a_predicted, b_predicted = regression.coef_[0], regression.intercept_ 
a_predicted, b_predicted
```

```python
predict = lambda x: x * a_predicted + b_predicted

plt.figure(figsize = (8,8))
plt.plot(X,y,'bo')
plt.plot((0,25), (predict(0),predict(25)), 'r-')
```

```python
pred = model.predict([[25]])
pred
```

```python
plt.figure(figsize = (8,8))
plt.plot(X,y,'bo')
plt.plot((0,25), (predict(0),predict(25)), 'r-')
plt.plot((25,),(pred[0],), 'go', markersize=10) #zaznaczenie naszego punktu pred
```

## Level 2 - Klasyfikacja
### KNeighboursClassifier

```python
import numpy as np
import matplotlib.pyplot as pit
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data" 
names = ('sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class')
df = pd.read_csv(url, names=names) 
df.head()
```

```python
df.groupby('Class').size()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

X = df.drop('Class',axis=1)
y = df['Class']

le = LabelEncoder()
y = le.fit_transform(y)
y
```

### Wizualizacja danych

```python
import seaborn as sns

sns.pairplot(df, hue = 'Class', size = 3, markers = ['o','s','D'])
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

plt.figure(figsize = (15,10))
parallel_coordinates(df, 'Class')
plt.title('Parallel Coordinates Plot', fontsize = 20, fontweight = 'bold')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = df.drop('Class', axis=1).values 
y = df['Class'].values

le = LabelEncoder() 
y = le.fit_transform(y)

fig = plt.figure(figsize=(10, 8)) 
ax = Axes3D(fig, elev=48, azim=134) 
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', s = X[:, 3]*50)
for name, label in [('Virginica', 0), ('Setosa', 1), ('Versicolour', 2)]: 
      ax.text3D(X[y == label, 0].mean(),
                X[y == label, 1].mean(),
                X[y == label, 2].mean(), name, 
                horizontalalignment='center',
                bbox=dict(alpha=.5, edgecolor='w', facecolor='w'),size=25)
        
        
ax.set_title("3D visualization", fontsize=40) 
ax.set_xlabel("Sepal Length [cm]", fontsize=25) 
ax.w_xaxis.set_ticklabels([]) 
ax.set_ylabel("Sepal Width [cm]", fontsize=25) 
ax.w_yaxis.set_ticklabels([]) 
ax.set_zlabel("Petal Length [cm]", fontsize=25)
plt.show()
```

### Korelacje

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(10,8))
ax = sns.heatmap(corr, annot = True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

### Podział danych

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Uczenie modelu

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

### Ocena modelu

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score

cf_matrix = confusion_matrix(y_test, y_pred)
cf_matrix
```

```python
plt.figure(figsize = (10,8))
plt.imshow(cf_matrix)
plt.grid(color = 'w', linestyle = '-', linewidth = 2)
plt.xticks([-0.5, 0.5, 1.5, 2.5])
plt.yticks([-0.5, 0.5, 1.5, 2.5])
plt.title('Confusion matrix')
plt.xlabel('Actual class')
plt.ylabel('Predicted class')
plt.show()
```

```python
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy of the model is ' + str(round(accuracy, 2)) + '%')
```
